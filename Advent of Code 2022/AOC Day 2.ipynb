```python
def RPS(x,y):
    global score
    
    #for a draw
    if x == y:
        if y == "X":
            score += 4
        elif y == "Y":
            score += 5
        elif y == "Z":
            score += 6
    
    #if player 1 selects rock
    elif x == "A":
        if y == "Y": #paper
            score += 8
        elif y == "Z": #scissors
            score += 3
    
    #player 1 selects paper
    elif x == "B":
        if y == "X": #rock
            score += 1
        elif y == "Z": #scissors
            score += 9
            
    #player 1 selects scissors
    elif x == "C":
        if y == "X": #rock
            score += 7
        elif y == "Y": #paper
            score += 2
            
    print(score)
    return score
            
```
